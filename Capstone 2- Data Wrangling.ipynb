```python
# creating a copy
sample_stats = educ_stats[['Country Name', 'Country Code', 'Indicator Name', '2014']].copy()
print(sample_stats.head())
```

```python
sample_stats = sample_stats.dropna(axis='rows')
```

```python
# Factor #1 - adj net secondary
x = sample_stats.loc[sample_stats['Indicator Name'] == 'Adjusted net enrolment rate, lower secondary, both sexes (%)']
```

```python
x.head()
```

```python
# adj net secondary dataframe 
adj_sec = pd.DataFrame(x)
adj_sec
```

```python
# factor #2 - tertiary edc 
f = sample_stats.loc[sample_stats['Indicator Name'] == 'Enrolment in tertiary education, all programmes, both sexes (number)']
f.head()
```

```python
# tertiary edc dataframe
ter_educ = pd.DataFrame(f)
ter_educ
```

```python
# merging adj enrollment to gradrates 
df1 = grad_rates.merge(adj_sec, how='left')
```

```python
df1
```

```python
df1 = df1.rename(columns={"2014" : "Adj net enrollment of lower secondary"})
```
